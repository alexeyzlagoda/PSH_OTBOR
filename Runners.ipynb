```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
import solver



r = 1
resolution = 10**(-4)
duration = 10**(0)*3
omegas = [5.0,2.0]
phis = [np.pi,np.pi/2]
k_arr = [[5,5],
         [5,5]]
TimeAxis, Phi_Axis, Omega_Axis = solver.Integrator(resolution, duration, omegas, phis, k_arr)
#Phi_Axis = np.transpose(Phi_Axis)
n = len(Phi_Axis[0])
#Phi = np.transpose(Phi)

'''
Phi = [
    [0, 2*np.pi],
    [np.pi, 2*np.pi],
    [0, np.pi]
]
'''
Colors = [
'green',
'red',
'blue'
]

figure, axes = plt.subplots()
# identify figures apart from axes

def frame (i):
    axes.clear()

    axes.set_xlim(-r * 1.1, r*1.1)
    axes.set_ylim(-r * 1.1, r*1.1)
    axes.set_aspect(1)
    phii = np.linspace(0, 2*np.pi, 100)
    x = np.cos(phii)
    y = np.sin(phii)
    axes.plot(x,y, color = 'black') 

    for j, phi in enumerate(Phi_Axis):
        x = np.cos(phi[i])
        y = np.sin(phi[i])
        axes.plot(x, y, marker = 'o', color = Colors[j], markersize = 10, zorder = 10)

frames_numbers = np.arange(n)
anime = FuncAnimation(figure, frame, frames = frames_numbers, interval  = 40)
anime.save('Runners.gif') 
display(anime)

```
