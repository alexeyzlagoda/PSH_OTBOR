```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
from tqdm import tqdm
import solver2 as solver
```

```python
r = 1
Colors = [
'green',
'red',
'blue'
]

resolution = 10**(-2)
duration = 10**(0)*10
omegas = [5.0,2.0]
phis = [np.pi,np.pi*0.75]
k_arr = [[-5,-5],
         [-5,-5]]
```

```python
TimeAxis, Phi_Axis, Omega_Axis = solver.Integrator(resolution, duration, omegas, phis, k_arr)
```

```python
n = len(Phi_Axis[0])
```

```python
 
plt.plot(TimeAxis,(Omega_Axis)[0], color = "red", label = "Omega 1")
plt.plot(TimeAxis,(Omega_Axis)[1], color = "blue", label = "Omega 2")
plt.title("    ")
plt.legend()
plt.savefig('Omegas.pdf')
plt.show()
plt.plot(TimeAxis,(Phi_Axis)[0], color = "red", label = "Phi 1")
plt.plot(TimeAxis,(Phi_Axis)[1], color = "blue", label = "Phi 2")
plt.title("    ")
plt.legend()
plt.savefig('Phis.pdf')
plt.show()
```

```python
figure, axes = plt.subplots()

def frame (i):
    axes.clear()

    axes.set_xlim(-r * 1.1, r*1.1)
    axes.set_ylim(-r * 1.1, r*1.1)
    axes.set_aspect(1)
    phii = np.linspace(0, 2*np.pi, 100)
    x = np.cos(phii)
    y = np.sin(phii)
    axes.plot(x,y, color = 'black') 

    for j, phi in enumerate(Phi_Axis):
        x = np.cos(phi[i])
        y = np.sin(phi[i])
        axes.plot(x, y, marker = 'o', color = Colors[j], markersize = 10, zorder = 10)
```

```python
frames_numbers = np.arange(n)
anime = FuncAnimation(figure, frame, frames = tqdm(frames_numbers[::10]), interval  = 1000 )
anime.save('Runners.gif') 
display(anime)
```
